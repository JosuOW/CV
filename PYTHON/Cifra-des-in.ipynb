```python
import tkinter as tk
import numpy as np
from tkinter import Entry, Label, Button
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Función para cifrar un mensaje utilizando RSA
def encrypt(message, public_key):
    n, e = public_key
    encrypted_message = [pow(ord(char), e, n) for char in message]
    return encrypted_message

# Función para descifrar un mensaje utilizando RSA
def decrypt(ciphertext, private_key):
    n, d = private_key
    decrypted_message = ''.join([chr(pow(char, d, n)) for char in ciphertext])
    return decrypted_message

# Generación de claves RSA (clave pública y privada)
def generate_rsa_keys(p, q):
    n = p * q
    phi = (p - 1) * (q - 1)
    e = 65537  # Se elige un valor común para e
    d = pow(e, -1, phi)  # Calculamos el inverso multiplicativo modular de e mod phi
    public_key = (n, e)
    private_key = (n, d)
    return public_key, private_key

# Función que se llama cuando se hace clic en el botón para cifrar
def cifrar_textos():
    # Obtener los textos ingresados por el usuario
    texto1 = entry_texto1.get()
    texto2 = entry_texto2.get()
    texto3 = entry_texto3.get()

    # Generar claves RSA
    p = 61
    q = 53
    public_key, private_key = generate_rsa_keys(p, q)

    # Cifrar los textos
    texto_cifrado1 = encrypt(texto1, public_key)
    texto_cifrado2 = encrypt(texto2, public_key)
    texto_cifrado3 = encrypt(texto3, public_key)

    # Actualizar las etiquetas con los textos cifrados
    label_resultado1.config(text=f"Texto 1 cifrado: {texto_cifrado1}")
    label_resultado2.config(text=f"Texto 2 cifrado: {texto_cifrado2}")
    label_resultado3.config(text=f"Texto 3 cifrado: {texto_cifrado3}")

# Función que se llama cuando se hace clic en el botón para descifrar
def descifrar_textos():
    # Obtener los textos cifrados desde las etiquetas
    global texto_descifrado1, texto_descifrado2, texto_descifrado3
    texto_cifrado1 = eval(entry_texto1.get())
    texto_cifrado2 = eval(entry_texto2.get())
    texto_cifrado3 = eval(entry_texto3.get())

    # Generar claves RSA
    p = 61
    q = 53
    _, private_key = generate_rsa_keys(p, q)

    # Descifrar los textos
    texto_descifrado1 = decrypt(texto_cifrado1, private_key)
    texto_descifrado2 = decrypt(texto_cifrado2, private_key)
    texto_descifrado3 = decrypt(texto_cifrado3, private_key)

    # Actualizar las etiquetas con los textos descifrados
    #label_resultado1.config(text=f"Texto 1 descifrado: {texto_descifrado1}")
   # label_resultado2.config(text=f"Texto 2 descifrado: {texto_descifrado2}")
   # label_resultado3.config(text=f"Texto 3 descifrado: {texto_descifrado3}")

#Generación matriz similitud

    documentos = [
        texto_descifrado1,
        texto_descifrado2,
        texto_descifrado3
    ]

    # Vectorización utilizando TF-IDF
    vectorizer = TfidfVectorizer()
    matriz_tfidf = vectorizer.fit_transform(documentos)

    # Calcula la matriz de similaridad de coseno
    matriz_similaridad = cosine_similarity(matriz_tfidf)

    print("Matriz de Similaridad:")
    print(matriz_similaridad)    

    def cholesky_simulation(corr_matrix, num_samples):
        # Descomposición de Cholesky
        L = np.linalg.cholesky(corr_matrix)
        # Generación de ruido incorrelacionado
        uncorrelated_noise = np.random.normal(size=(len(corr_matrix), num_samples))
        # Aplicar la descomposición de Cholesky al ruido
        correlated_noise = np.dot(L, uncorrelated_noise)
        return correlated_noise

    def calculate_uncertainty(original_matrix, num_samples, num_iterations):
        uncertainties = []

        for _ in range(num_iterations):
            # Generar matriz de similitud simulada
            simulated_matrix = cholesky_simulation(original_matrix, num_samples)

            # Calcular la incertidumbre
            uncertainty = np.std(simulated_matrix, axis=1)
            uncertainties.append(uncertainty)

        uncertainties = np.array(uncertainties)
        average_uncertainty = np.mean(uncertainties, axis=0)

        return average_uncertainty

    # Ingreso de la matriz de similitud por consola
    original_matrix = matriz_similaridad
    # Parámetros de la simulación
    num_samples = 1000
    num_iterations = 100

    # Calcular la incertidumbre mediante Monte Carlo y descomposición de Cholesky
    uncertainty_result = calculate_uncertainty(original_matrix, num_samples, num_iterations)

    # Mostrar resultados
    print("\nMatriz de Similitud Original:")
    print(original_matrix)
    print("\nIncertidumbre Estimada:")
    print(uncertainty_result)    
    arreglo = np.array(uncertainty_result)
    indice_maximo = np.argmax(arreglo)
    texto=documentos[indice_maximo] 
    label_resultadotext.config(text=f"Contraseña en: {texto}")
# Crear la ventana principal
ventana = tk.Tk()
ventana.title("Cifrado y Descifrado RSA para detectar Contraseña ")

# Crear etiquetas y campos de entrada para los textos cifrados
Label(ventana, text="Mensaje 1:").pack()
entry_texto1 = Entry(ventana)
entry_texto1.pack()

Label(ventana, text="Mensaje 2:").pack()
entry_texto2 = Entry(ventana)
entry_texto2.pack()

Label(ventana, text="Mensaje 3:").pack()
entry_texto3 = Entry(ventana)
entry_texto3.pack()

# Botones para cifrar y descifrar los textos
Button(ventana, text="Cifrar", command=cifrar_textos).pack()
label_resultado1 = Label(ventana, text="")
label_resultado1.pack()

label_resultado2 = Label(ventana, text="")
label_resultado2.pack()

label_resultado3 = Label(ventana, text="")
label_resultado3.pack()

Button(ventana, text="Contraseña en Texto", command=descifrar_textos).pack()

# Etiquetas para mostrar los textos descifrados


label_resultadotext = Label(ventana, text="")
label_resultadotext.pack()
# Iniciar el bucle principal de la interfaz gráfica
ventana.mainloop()

```

```python
import numpy as np

# Función para cifrar un mensaje utilizando RSA
def encrypt(message, public_key):
    n, e = public_key
    encrypted_message = [pow(ord(char), e, n) for char in message]
    return encrypted_message

# Función para descifrar un mensaje utilizando RSA y Gauss-Jordan
def decrypt(ciphertext, private_key):
    n, d = private_key
    decrypted_message = [pow(char, d, n) for char in ciphertext]
    return ''.join(chr(char) for char in decrypted_message)

# Generación de claves RSA (clave pública y privada)
def generate_rsa_keys(p, q):
    n = p * q
    phi = (p - 1) * (q - 1)
    e = 65537  # Se elige un valor común para e
    d = pow(e, -1, phi)  # Calculamos el inverso multiplicativo modular de e mod phi
    public_key = (n, e)
    private_key = (n, d)
    return public_key, private_key

# Generamos un par de claves RSA (clave pública y privada)
p = 61
q = 53
public_key, private_key = generate_rsa_keys(p, q)

# Mensaje a cifrar
message = "Hola mundo, estoy probando el cifrado"

# Cifrar el mensaje utilizando la clave pública
ciphertext = encrypt(message, public_key)

print("Mensaje cifrado:", ciphertext)

# Implementación del método de descifrado utilizando Gauss-Jordan
def decrypt_gauss_jordan(ciphertext, private_key):
    n, d = private_key
    decrypted_message = []

    # Convertimos los números cifrados en una matriz
    encrypted_matrix = np.array(ciphertext).reshape(len(ciphertext), 1)

    # Creamos la matriz de clave privada como una matriz diagonal
    private_key_matrix = np.diag([d] * len(ciphertext))

    # Aplicamos el método de Gauss-Jordan para resolver el sistema de ecuaciones lineales
    decrypted_matrix = np.linalg.solve(private_key_matrix, encrypted_matrix)

    # Convertimos la matriz resultante en el mensaje descifrado
    decrypted_message = [int(round(x[0])) for x in decrypted_matrix]

    return ''.join(chr(char) for char in decrypted_message)

# Descifrar el mensaje utilizando el método de Gauss-Jordan
decrypted_message = decrypt_gauss_jordan(ciphertext, private_key)

print("Mensaje descifrado:", decrypted_message)
```
