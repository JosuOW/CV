```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Función para la ecuación diferencial
def ecuacion(y, t):
    return eval(equation)

# Método de Euler mejorado aproximación de segundo orden
def euler_mejorado(h, y0, t0, tn):
    start_time = time.time()  # Medir el tiempo de inicio
    n = int((tn - t0) / h) + 1
    t = np.linspace(t0, tn, n)
    y = np.zeros(n)
    y[0] = y0

    for i in range(1, n):
        k1 = h * ecuacion(y[i-1], t[i-1])
        k2 = h * ecuacion(y[i-1] + k1, t[i])
        y[i] = y[i-1] + 0.5 * (k1 + k2)

    end_time = time.time()  # Medir el tiempo de finalización
    tiempo = end_time - start_time
    return t, y, tiempo

# Método de Runge-Kutta de cuarto orden
def runge_kutta(h, y0, t0, tn):
    start_time = time.time()  # Medir el tiempo de inicio
    n = int((tn - t0) / h) + 1
    t = np.linspace(t0, tn, n)
    y = np.zeros(n)
    y[0] = y0

    for i in range(1, n):
        k1 = h * ecuacion(y[i-1], t[i-1])
        k2 = h * ecuacion(y[i-1] + 0.5 * k1, t[i-1] + 0.5 * h)
        k3 = h * ecuacion(y[i-1] + 0.5 * k2, t[i-1] + 0.5 * h)
        k4 = h * ecuacion(y[i-1] + k3, t[i-1] + h)
        y[i] = y[i-1] + (k1 + 2*k2 + 2*k3 + k4) / 6

    end_time = time.time()  # Medir el tiempo de finalización
    tiempo = end_time - start_time
    return t, y, tiempo

# Función mostrar resultados
def mostrar_resutados(t, y):
    print("\nResultados:")
    for i in range(len(t)):
        print(f"t = {t[i]:.2f}, y = {y[i]:.4f}")

# Obtener la ecuación diferencial y los parámetros por consola
equation = input("Ingrese la ecuación diferencial en términos de y y t, ej. 'y - t**2 + 1': ")

h = float(input("Ingrese el tamaño del paso h ej. '0.2': "))
y0 = float(input("Ingrese el valor inicial y0 ej. '0': "))
t0 = float(input("Ingrese el tiempo inicial t0 ej. '0': "))
tn = float(input("Ingrese el tiempo final tn ej. '2': "))

# Obtener resultados con ambos métodos
t_euler, y_euler, tiempo_euler = euler_mejorado(h, y0, t0, tn)
print("\nMétodo de Euler mejorado")
mostrar_resutados(t_euler, y_euler)
print(f"Tiempo de ejecución: {tiempo_euler} segundos")

t_rk, y_rk, tiempo_rk = runge_kutta(h, y0, t0, tn)
print("\nMétodo de Runge-Kutta")
mostrar_resutados(t_rk, y_rk)
print(f"Tiempo de ejecución: {tiempo_rk} segundos")

# Graficar resultados
plt.plot(t_euler, y_euler, label='Euler Mejorado')
plt.plot(t_rk, y_rk, label='Runge-Kutta')
plt.legend()
plt.title('Comparación entre Euler Mejorado y Runge-Kutta')
plt.xlabel('Tiempo (t)')
plt.ylabel('y(t)')
plt.show()

```
